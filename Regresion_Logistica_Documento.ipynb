3.Preparacion de datos
>- data['PerformanceScore'] =  pd.qcut(data['PerformanceScore'], q=6, labels=[0, 1, 2, 3, 4, 5])
>- x = data[[ 'Age', 'YearsAtCompany', 'PerformanceScore', 'Salary']]
>- y = data['PerformanceScore']
>- x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
>- scaler = StandardScaler()
>- x_train_scaled = scaler.fit_transform(x_train)
>- x_test_scaled = scaler.transform(x_test)
>>- data['PerformanceScore'] =  pd.qcut(data['PerformanceScore'], q=6, labels=[0, 1, 2, 3, 4, 5]): Esto sirve para dividir los valores de "PerformanceScore" en 6 intervalos (o “bins”) utilizando la función pd.qcut(). Cada intervalo se asigna a una etiqueta categórica (0, 1, 2, 3, 4 o 5).
>>- x: define las caracteristicas que se utilizaran para hacer las predicciones tales como "edad, años en compañia, puntuacion de rendimiento y salario".
>>- y: Define la variable objetivo la cual es la columna PerformanceScore donde 1 significa que tiene un rendimiento alto y 0 un rendimiento bajo esta es la columna que se va a predecir.
>>- train_test_split: Divide los datos en dos subconjuntos:
>>- x_train y y_train: El 80% de los datos que se usarán para entrenar el modelo.
>>- x_test y y_test: El 20% restante se usará para probar el modelo. test_size=0.2: Indica que el 20% de los datos será usado para pruebas.
>>- random_state=42: Asegura que la división sea reproducible, es decir, cada vez que corras el código, obtendrás la misma división de datos.
>>- scaler = StandardScaler(): Inicializa un objeto StandardScaler para escalar los datos
>>- fit_transform(): Calcula los parametros de escalado (media y desviacion estandar) en el conjunto de entrenamiento y aplica la transformación
>>- Transform(): Aplica la misma transformacion al conjunto de prueba. Escalamos los datos que tengan una media de 0 y desviación estandar de 1, lo que ayuda a que el modelo converja mejor.
4.Entrenamiento del modelo
>- logistic_model = LogisticRegression()
>- logistic_model.fit(x_train_scaled, y_train)
>>- logistic_model = LogisticRegression(): Inicializa el modelo de regresión logistica
>>- logistic_model.fit(x_train_scaled, y_train): Entrena el modelo usando el conjunto de entrenamiento escalado y las etiquetas.
5.Realizar predicciones
>- y_pred = logistic_model.predict(x_test_scaled)
>>- y_pred = logistic_model.predict(x_test_scaled): Contiene las predicciones del modelo en el conjunto de prueba escalado donde el modelo predice si tendra buen desempeño 1 o bajo 0.
6.Evaluación del modelo
>-conf_matrix = confusion_matrix(y_test, y_pred)
>-plt.figure(figsize=(8,6))
>-sns.heatmap(conf_matrix, annot=True, fmt='d', cmap ='Blues', cbar=False)
>-plt.xlabel('Predicted')
>-plt.ylabel('Actual')
>-plt.title('Confusion Matrix')
>-plt.show()
>>- conf_matrix = confusion_matrix(y_test, y_pred): Calcula la matriz de confusión que compara las predicciones del modelo (y_pred) con los valores reales (y_test). la matriz muestra cuantos resultados fueron predecidos correctamente y cuantos fueron incorrectos.
>>- plt.figure(figsize=(8,6)): establece el tamaño del grafico.
>>- sns.heatmap(conf_matrix, annot=True, fmt='d', cmap ='Blues', cbar=False): Crea un mapa de calor para visualizar la matriz de confusión.
>>-plt.xlabel(), plt.ylabel(), plt.tittle(): añade etiquetas al grafico junto a un titulo.
>>- plt.show(): Muestra el grafico.

7.Reporte de clasificación
>- print(classification_report(y_test, y_pred))
>>- utiliza la función classification_report() de sklearn.metrics para generar un informe detallado sobre el rendimiento del modelo de clasificación. Este informe incluye varias metricas de evaluación clave.
>>>- 1.Precisión: mide que porcentaje de las predicciones positivas del modelo son correctas
>>>- 2.Sensibilidad: Mide que porcentaje de los verdaderos casos positivos fueron detectados correctamente por el modelo.
>>>- 3.Puntuación F1: Es la media armonica entre la precision y sensibilidad. Es util cuando hay un desequilibrio entre clases y queremos equilibrar la precision y el recall.
>>>- 4.Soporte:Indica el numero de ocurrencias reales de cada clase en los datos de prueba.
